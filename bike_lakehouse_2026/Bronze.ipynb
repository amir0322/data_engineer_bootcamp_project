### read from csv file

```python
df =spark.read.format("csv").option("header","true").option("inferSchema","true")\
    .load("/Volumes/data_engineering_project/bronze/source_system/source_crm/cust_info.csv")

df.display()
```

### write to Bronze Layer

```python
df.write.mode("overwrite").saveAsTable("data_engineering_project.bronze.crm_cust_info")
```
