### The Transformation Logic

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY customer_id) AS customer_key,
    customer_id,
    customer_number,
    first_name,
    last_name,
    marital_status,
    gender,
    created_date AS create_date
FROM data_engineering_project.silver.crm_customers
"""
df = spark.sql(query)
```

```python
df.limit(10).display()
```

### Writing Gold Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("data_engineering_project.gold.dim_customers")
```
